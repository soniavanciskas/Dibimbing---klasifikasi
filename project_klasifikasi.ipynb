```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
# load dataset
cancer = datasets.load_breast_cancer()

x = cancer.data
y = cancer.target
y
```

```python
# ubah menjadi dataframe
dfx= pd.DataFrame(x, columns=cancer.feature_names)
dfy = pd.Series(y, name='target')
```

```python
#gabung dataframe
df = pd.concat([dfx, dfy], axis=1)
df
```

<h2> Data preprocessing

```python
#penjelasan mengenai class yang ada di dalam data
df.info()
```

```python
#cek data hilang
df.isna().sum()
```

```python
# cek data duplikat
df.duplicated().sum()
```

```python
# cek outlier dalam data
for kol in dfx :
    plt.figure(figsize = (20,2))
    sns.boxplot(df, x=kol)
```

```python
# cek outlier dengan detail
q1 = df.quantile(0.25, numeric_only=True)
q3 = df.quantile(0.75, numeric_only=True)
iqr = q3 - q1

bawah = q1 - (1.5 * iqr)
atas = q3 + (1.5 * iqr)

out = (df < bawah ) | (df > atas)

for kol in out.columns:
    print (f'kolom  {kol}')
    print (out[kol].value_counts())
    print('-' * 30)
```

```python
# penanganan nilai outlier dengan winsorize
from scipy.stats.mstats import winsorize

for kol in out.columns:
    if out[kol].any():
        lower_bound = bawah[kol]
        upper_bound = atas[kol]

        lower_limit = (df[kol] < lower_bound).mean()
        upper_limit = (df[kol] > upper_bound).mean()

        df[kol] = winsorize(df[kol], limits=(lower_limit, upper_limit))
        plt.figure(figsize=[20,2])
        sns.boxplot(data=df, x=df[kol])
```

```python

# karena distribusi data memiliki rentang yang yang sangat berbeda, data akan distandarisasi
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df.describe()
```

<h2> Modelling

```python
# split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dfx, dfy, test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

image.png

<h3> Model Decision Tree


```python
from sklearn.tree import DecisionTreeClassifier
decitree = DecisionTreeClassifier(random_state=42)
decitree.fit(x_train, y_train)
```

```python
ypred_decitree = decitree.predict(x_test)
ypred_decitree
```

<h3> Model Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)
```

```python
ypred_rf = rf.predict(x_test)
ypred_rf
```

<h3> Model Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=42, max_iter=500)
logreg.fit(x_train, y_train)
```

```python
ypred_logreg = logreg.predict(x_test)
ypred_logreg
```

<h3> Model Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
ypred_nb = nb.predict(x_test)
ypred_nb
```

<h2> Evaluasi model

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

akurasi_decitree = accuracy_score(y_test, ypred_decitree)
akurasi_rf = accuracy_score(y_test, ypred_rf)
akurasi_nb = accuracy_score(y_test, ypred_nb)
akurasi_logreg = accuracy_score(y_test, ypred_logreg)

cf_decitree = confusion_matrix(y_test, ypred_decitree)
cf_rf = confusion_matrix(y_test, ypred_rf)
cf_nb = confusion_matrix(y_test, ypred_nb)
cf_logreg = confusion_matrix(y_test, ypred_logreg)

cr_decitree = classification_report(y_test, ypred_decitree)
cr_rf = classification_report(y_test, ypred_rf)
cr_nb = classification_report(y_test, ypred_nb)
cr_logreg = classification_report(y_test, ypred_logreg)
```
